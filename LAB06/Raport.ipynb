## Vizualizim — Krahasimi i Metodave të Matjes

Ndërtojmë një grafik me shtylla që krahason matjet e të njëjtave objekte me vizore, kaliber (pa nonius) dhe kaliber (me nonius), të ruajtur në `figs/`.

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
Path("figs").mkdir(exist_ok=True)

objektet = ["100 g", "20 g", "10 g"]
vizore  = [32.0, 22.0, 14.0]   # mm (3.2/2.2/1.4 cm)
kaliber = [33.1, 22.2, 15.4]   # mm
nonius  = [33.12, 22.24, 15.40]# mm

x = np.arange(len(objektet)); w = 0.25
fig, ax = plt.subplots(figsize=(8,4.5))
ax.bar(x-w, vizore,  w, label="Vizore")
ax.bar(x,   kaliber, w, label="Kaliber (pa nonius)")
ax.bar(x+w, nonius,  w, label="Kaliber (me nonius)")
ax.set_xticks(x); ax.set_xticklabels(objektet)
ax.set_xlabel("Objekti"); ax.set_ylabel("Përmasa (mm)")
ax.set_title("Krahasimi i metodave të matjes")
ax.legend(); ax.grid(axis="y", alpha=0.3); fig.tight_layout()
fig.savefig("figs/krahasimi_matjeve.png", dpi=150); plt.show()
```
