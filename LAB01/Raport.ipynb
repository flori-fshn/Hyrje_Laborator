## Verifikim i Mjedisit — Grafiku i Parë

Për të verifikuar se Conda + Jupyter + matplotlib punojnë, ndërtojmë një grafik të thjeshtë dhe e ruajmë në `figs/`.

````python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
Path("figs").mkdir(exist_ok=True)

x = np.linspace(0, 2*np.pi, 200)
plt.figure(figsize=(8,4))
plt.plot(x, np.sin(x), label="sin(x)")
plt.plot(x, np.cos(x), label="cos(x)")
plt.xlabel("x (rad)"); plt.ylabel("f(x)")
plt.title("Verifikim: matplotlib punon në Jupyter")
plt.legend(); plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig("figs/demo_plot.png", dpi=150)
plt.show()
````
