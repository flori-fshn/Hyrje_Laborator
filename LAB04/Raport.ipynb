### Shembull: **Fit linear**

```python
import numpy as np

x = np.array([1,2,3,4])
y = np.array([2,4,6,8])

coeff = np.polyfit(x,y,1)

slope = coeff[0]
intercept = coeff[1]
```

```python
y = slope * x + intercept
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4])
y = np.array([2.1,4.2,6.1,8.0])

coeff = np.polyfit(x,y,1)

fit = coeff[0]*x + coeff[1]

plt.scatter(x,y,label="Data")
plt.plot(x,fit,label="Fit")

plt.xlabel("x")
plt.ylabel("y")

plt.legend()
plt.show()
```

## Vizualizim — Funksione me Python/NumPy

Demonstrojmë përdorimin e NumPy dhe Matplotlib duke vizatuar disa funksione bazë dhe duke e ruajtur figurën në `figs/`.

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
Path("figs").mkdir(exist_ok=True)

x = np.linspace(0, 5, 200)
plt.figure(figsize=(8,4.5))
for p, lbl in [(1,"x"),(2,"x²"),(0.5,"√x")]:
    plt.plot(x, x**p, label=f"y = {lbl}")
plt.xlabel("x"); plt.ylabel("y")
plt.title("Funksione bazë në Python")
plt.legend(); plt.grid(alpha=0.3); plt.tight_layout()
plt.savefig("figs/funksionet.png", dpi=150)
plt.show()
```
