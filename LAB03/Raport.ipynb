- ##  Cfare eshte Markdown?
---

**Markdown** eshe nje gjuhe e thjeshte për formatimin e tekstit qe perdoret per te shkruar dokumente të strukturuara.

Në Jupyter Notebook, qelizat Markdown perdoren per:
- pershkrimin e eksperimentit
- shpjegimin e teorise
- interpretimin e rezultateve
- shkrimin e formulave matematikore

Markdown eshte shumë i thjeshtë sepse perdor disa simbole baze per te strukturuar tekstin.

**Ne kete laborator do te perdorim Markdown per:**
- Tituj dhe nenseksione
- Tekst te theksuar (bold,italic)
- Tabela
- Formula matematike me LaTex
- Shtimin e figurave
- Listat dhe code blocks

**Mjetet qe perdorem:**

Per kete laborator kemi perdorur:
- **JupyterLab** i instaluar me **Anaconda/Conda++**
- **Python kernel** per shembujt e kodeve
- Markdown per formatimin e raporteve
- Skedarin e figures ('example.png') per shembullin visual

| Sintaksa | Rezultati |

## Vizualizim — Përfshirja e një Figure në Markdown

Markdown-i lejon përfshirjen e figurave. Më poshtë krijojmë një grafik shembull dhe e ruajmë në `figs/` për ta referuar në raport.

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
Path("figs").mkdir(exist_ok=True)

x = np.linspace(-3, 3, 200)
plt.figure(figsize=(8,4))
plt.plot(x, np.exp(-x**2), color="#8e44ad")
plt.fill_between(x, np.exp(-x**2), alpha=0.2, color="#8e44ad")
plt.xlabel("x"); plt.ylabel("exp(-x²)")
plt.title("Shembull grafiku për përfshirje në Markdown")
plt.grid(alpha=0.3); plt.tight_layout()
plt.savefig("figs/shembull_grafik.png", dpi=150)
plt.show()
```
